```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('spam_assassin.csv')
```

```python
def clean_email(email):
    clean_email = []
    for i in email:
        if not i.isalnum():
            i = i.replace(i, ' ')
            clean_email.append(i)
        else:
            i = i.lower()
            clean_email.append(i)
    clean_email = "".join(clean_email)
    return clean_email
```

```python
def convert_email_in_dict(email):
    word_freq = {}
    for word in email.split():
        if word in word_freq:
            word_freq[word] +=1
        else:
            word_freq[word] = 1
    return word_freq
```

```python
def words_to_df(word_freq):
    df_word = pd.DataFrame.from_dict(word_freq, orient = 'index').fillna(0)
    df_word = df_word.T
    return df_word
```

```python
all_dfs = []

for i in range(len(df)):

    # ✅ progress checkpoint
    if i % 100 == 0:
        print(f"Processed {i} emails")

    clean = clean_email(df.iloc[i, 0])
    word_freq = convert_email_in_dict(clean)
    all_dfs.append(word_freq)

final_dfs = pd.DataFrame(all_dfs).fillna(0)
```

```python
final_dfs
```

```python
target = df["target"]
final_dfs["Target"] = target
final_dfs["Target"]
```

```python
X = final_dfs.drop("Target",axis = 1)
y = final_dfs["Target"]
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
splitter = StratifiedShuffleSplit(n_splits= 1, test_size= 0.2, random_state= 42)

for train_index,test_index in splitter.split(final_dfs,final_dfs['Target']):
    X_train = final_dfs.drop(columns = ["Target"]).iloc[train_index]
    y_train = final_dfs["Target"].iloc[train_index]
    X_test = final_dfs.drop(columns=["Target"]).iloc[test_index]
    y_test = final_dfs["Target"].iloc[test_index]
```

```python
for train_index_val, test_index_val in splitter.split(X_train,y_train):
    X_train_final = X_train.iloc[train_index_val]
    y_train_final = y_train.iloc[train_index_val]

    X_val = X_train.iloc[test_index_val]
    y_val = y_train.iloc[test_index_val]
```

```python
print(X_train.shape, y_train.shape, X_val.shape, y_val.shape, X_test.shape, y_test.shape)
```

<h1>Logistic Regression</h1>

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train_final,y_train_final)
```

```python
y_pred = model.predict(X_val)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_val,y_pred))
print(confusion_matrix(y_val,y_pred))
```

```python
y_test_pred = model.predict(X_test)
```

```python
print(classification_report(y_test,y_test_pred))
print(confusion_matrix(y_test, y_test_pred))
```
